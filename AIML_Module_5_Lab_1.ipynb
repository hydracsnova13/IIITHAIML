# K-Means

K-Means algorithm is a centroid based clustering algorithm where the sum of distances of points from the centroid of each cluster is minimized. The final output is a set of K clusters .ie. the cluster assigned to each point and the K centroids of the clusters.

### The Algorithm
> 1.   Initialize K centroids to K points randomly and set each point's initial cluster as the centroid closest to it.
2.   While the clusers are changing - \\
a. Assign the new centroids as the centroids of the points which have the same assigned cluster. \\
b. Assign the new clusters to the points as the closest centroid.
3. Return the assignments and the centroids.





### Implementation of KMeans and testing on Synthetic Datasets

```python
# Importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generating Synthetic Data

k=2
color=["darkturquoise","darkorange","teal","darkviolet","tomato", "yellowgreen","hotpink","gold"]

X,_=make_blobs(n_samples=500, n_features=2, centers=k, random_state=10)
print("Shape = " + str(X.shape))

print("Sample:")
print(X[:5])
```

```python
#DEFINE ALL THE FUNCTIONS & VARIABLES WE NEED TO DO CLUSTERING
#Create an empty dictionary to store the cluster name, centroid and points associated with the cluster.
clusters={}

def initializeClusterCentroids():
  #For each cluster, initialize its properties
  for i in range(k):
      center = np.random.randint(-8, 10, size= (X.shape[1],), )
      points = []

      cluster = {
          'center' : center,
          'points' : points,
          'color' : color[i]
      }
      clusters[i] = cluster #looping over clusters dictionary

#Calculate euclidean distance between v1 and v2
def distance(v1, v2):
    return np.sum((v2-v1)**2)**0.5

def assignPointsToCluster(clusters, X):
    for ix in range(X.shape[0]):#iterate for all datapoints

        #Foe each datapoint, find the distance to k centroids
        distance_of_i = []
        for kx in range(k):
            d = distance(X[ix], clusters[kx]['center']) #we have find distance b/w
            #all the point with all the five (k) cluster centers
            distance_of_i.append(d)

        #Based on the distances to k centroid, pick which cluster to assign it to
        #here we want find the minimum distance of cluster centers
        cluster_to_choose = np.argmin(distance_of_i)
        clusters[cluster_to_choose]['points'].append(X[ix])

#STEP 3: UPDATE CLUSTER CENTROIDS
def updateCluster(clusters):
    for kx in range(k):
        cluster_points = clusters[kx]['points']

        cluster_points = np.array(cluster_points)
        #here pts are array of list but we want array of array so

        if len(cluster_points)>0:
            new_center = np.mean(cluster_points, axis=0) #axis is along rows so will
            #find mean of all the feature seprately it will give 2,0 when we have cluster pts 50,2
            clusters[kx]['center'] = new_center
            clusters[kx]['points'] = [] # clear the points in a cluster list (emptying pts)
            #as we have to do step 2 after step 3 again till not converge
```

```python
#We need to visualize this too. Lets define a function that can do that!
def plotclusters(clusters, plot):
    for kx in range(k):
        cluster_points = clusters[kx]['points']
        cluster_color = clusters[kx]['color']
        cluster_center = clusters[kx]['center']
        cluster_points = np.array(cluster_points)

        #plotting points associated /nearest to the cluster centers
        if len(cluster_points) > 0:
          plot.scatter(cluster_points[:, 0], cluster_points[:, 1], s = 2, c = cluster_color)

        plot.scatter(cluster_center[0], cluster_center[1], s = 250, c = cluster_color, marker="o")
```

We can now run and visualize how the clusters evolve.

```python
fig, axs = plt.subplots(2, 3)
fig.set_size_inches(15, 10)
axs[0, 0].scatter(X[:,0],X[:,1], s = 2, c = 'black')
axs[0, 0].set_title('Dataset')

np.random.seed(1)
initializeClusterCentroids()

#Now, lets plot the centroids it has initalized
axs[0, 1].set_title('Centroid Initialization')
axs[0, 1].scatter(X[:, 0], X[:, 1], s = 2, c = 'black')
for i in range(k):#for all the clusters
    center = clusters[i]['center'] #its cluster's center
    axs[0, 1].scatter(center[0], center[1], c=clusters[i]['color'], s=250, marker="o")

#CLUSTERING EPOCH 1
axs[0, 2].set_title('Iteration 1')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[0, 2])
updateCluster(clusters)

#CLUSTERING EPOCH 2
axs[1, 0].set_title('Iteration 2')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[1, 0])
updateCluster(clusters)

#CLUSTERING EPOCH 3
axs[1, 1].set_title('Iteration 3')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[1, 1])
updateCluster(clusters)

#CLUSTERING EPOCH 4
axs[1, 2].set_title('Iteration 4')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[1, 2])
updateCluster(clusters)

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

### KMeans on Digits Dataset (KMeans for labelling)

Here we will attempt to use k-means to try to identify similar digits without using the original label information; this might be similar to a first step in extracting meaning from a new dataset about which you don't have any a priori label information.

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits.data.shape
```

```python
from sklearn.cluster import KMeans

# Perform KMeans
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

```python
# We can now see how the centers of the clusters formed are

fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

Because k-means knows nothing about the identity of the cluster, the 09 labels may be permuted. We can fix this by matching each learned cluster label with the true labels found in them:

```python
from scipy.stats import mode

labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
```

Now we can check how accurate our unsupervised clustering was in finding similar digits within the data:

```python
from sklearn.metrics import accuracy_score
accuracy_score(digits.target, labels)
```

**Hence clustering algorithms like KMeans are highly effective for identifying labels for groups apriori and even a simple KMeans acheives a ~80% accuracy on digit classification.**

The confusion matrix for this looks like:

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()

mat = confusion_matrix(digits.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

### KMeans on Iris Dataset (Elbow Method)

We will perform KMeans on the famous Iris Dataset which has four features: sepal length, sepal width, petal length, and petal width. The fifth column is for species, which holds the value for these types of plants.

```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from sklearn import datasets
```

```python
#import the dataset
iris = datasets.load_iris()
# print(iris.data[:10])
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
x = df.iloc[:, [0,1,2,3]].values
df.head(5)
```

Let us start with assigning the initial number of clusters as k=5.

```python
kmeans5 = KMeans(n_clusters=5)
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)

kmeans5.cluster_centers_
```

How do we choose the optimal number of clusters? Theres a method called the Elbow method, which is designed to help find the optimal number of clusters in a dataset. So lets use this method to calculate the optimum value of k. To implement the Elbow method, we need to create some Python code (shown below), and well plot a graph between the number of clusters and the corresponding error value.

This graph generally ends up shaped like an elbow, hence its name:

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)

import matplotlib.pyplot as plt
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

The output graph of the Elbow method shows that the shape of elbow is approximately formed at k=3. The optimal value of k is between 2 and 4, as the elbow-like shape is formed at k=3 in the above graph.

```python
kmeans3 = KMeans(n_clusters=3)
y_kmeans3 = kmeans3.fit_predict(x)
print(y_kmeans3)

kmeans3.cluster_centers_
```

We can now visualize the clustering in two dimensions instead of the original four.

```python
plt.scatter(x[:,0], x[:,1], c=y_kmeans3, cmap='rainbow')
```

1.Is feature scaling essential for KMeans as it is for most ML algos? Explain.

Yes, feature scaling is crucial for KMeans. KMeans clustering is based on distance metrics like Euclidean distance to form clusters. If the features are not on the same scale, the algorithm might end up giving undue importance to some features over others. For instance, if one feature ranges between 0-1 and another between 100-1000, the algorithm will be dominated by the larger range feature, potentially leading to skewed or incorrect clustering. Scaling ensures that each feature contributes equally to the distance calculations.

2.What are ways to prevent initialization variation in KMeans?

The most common approach to mitigate initialization variation in KMeans is to use the KMeans++ algorithm for initializing centroids. KMeans++ selects initial cluster centers in a smart way to speed up convergence.
Another approach is to run the KMeans algorithm multiple times with different random initializations and choose the one with the lowest SSE (sum of squared errors). This is often done automatically in many implementations of KMeans.
Some algorithms also use hierarchical clustering methods to determine initial centroids.

3.What is the training and testing complexity of KMeans?

Training Complexity: The training complexity of KMeans is O(n * k * i * d), where 'n' is the number of data points, 'k' is the number of clusters, 'i' is the number of iterations, and 'd' is the number of attributes. Since KMeans often requires several iterations to converge and operates over all data points and features, it can be computationally intensive, particularly with large datasets and a high number of dimensions.

Testing Complexity: KMeans doesn't have a traditional testing phase like supervised learning algorithms. Once the centroids of the clusters are determined after the training (fitting) phase, assigning a new data point to a cluster is a matter of calculating its distance to each centroid, which is a relatively fast operation (O(k * d), where 'k' is the number of clusters and 'd' is the number of attributes). However, the interpretability of these assignments depends on how well the model has been trained.

# Hierarchical Clustering

KMeans is an iterative process. It will keep on running until the centroids of newly formed clusters do not change or the maximum number of iterations are reached.

But there are certain challenges with K-means. It always tries to make clusters of the same size. Also, we have to decide the number of clusters at the beginning of the algorithm. Ideally, we would not know how many clusters should we have, in the beginning of the algorithm and hence it a challenge with K-means.

This is a gap hierarchical clustering bridges with aplomb. It takes away the problem of having to pre-define the number of clusters. Sounds like a dream! So, lets see what hierarchical clustering is and how it improves on K-means.

## Agglomerative Hierarchical Clustering

In this technique, initially each data point is considered as an individual cluster. At each iteration, the similar clusters merge with other clusters until one cluster or K clusters are formed.

The basic algorithm of Agglomerative is straight forward.

    Compute the proximity matrix
    Let each data point be a cluster
    Repeat: Merge the two closest clusters and update the proximity matrix
    Until only a single cluster remains

To understand better lets see a pictorial representation of the Agglomerative Hierarchical clustering Technique. Lets say we have six data points {A,B,C,D,E,F}.

    Step- 1: In the initial step, we calculate the proximity of individual points and consider all the six data points as individual clusters as shown in the image below.

image.png

    Step- 2: In step two, similar clusters are merged together and formed as a single cluster. Lets consider B,C, and D,E are similar clusters that are merged in step two. Now, were left with four clusters which are A, BC, DE, F.
    Step- 3: We again calculate the proximity of new clusters and merge the similar clusters to form new clusters A, BC, DEF.
    Step- 4: Calculate the proximity of the new clusters. The clusters DEF and BC are similar and merged together to form a new cluster. Were now left with two clusters A, BCDEF.
    Step- 5: Finally, all the clusters are merged together and form a single cluster.

You cut the dendrogram tree with a horizontal line at a height where the line can traverse the maximum distance up and down without intersecting the merging point.

## Divisive Hierarchical Clustering

In Divisive or DIANA(DIvisive ANAlysis Clustering) is a top-down clustering method where we assign all of the observations to a single cluster and then partition the cluster to two least similar clusters. Finally, we proceed recursively on each cluster until there is one cluster for each observation. So this clustering approach is exactly opposite to Agglomerative clustering.

image.png

There is evidence that divisive algorithms produce more accurate hierarchies than agglomerative algorithms in some circumstances but is conceptually more complex.

In both agglomerative and divisive hierarchical clustering, users need to specify the desired number of clusters as a termination condition(when to stop merging).

Divisive clustering methods are fairly complicated to implement, so we will not do that here. You can check out open source implementations like https://github.com/div338/Divisive-Clustering-Analysis-Program-DIANA- for further details.

1.Need for Hierarchical Clustering and Its Advantages over KMeans:

Hierarchical clustering is used when the data requires a hierarchy of clusters, which means it builds a hierarchy or a tree of clusters. Its particularly useful in fields like biology (for taxonomy) or linguistics (for understanding language structures), where such hierarchy is meaningful.

Advantages over KMeans:

-No Need to Specify Number of Clusters: Unlike KMeans, hierarchical clustering doesn't require you to specify the number of clusters upfront.

-The hierarchical tree (dendrogram) produced can be cut at the desired number of clusters.

-Captures Hierarchical Relationships: It provides more information about the dataset by creating a hierarchy, which can be insightful.

-Flexibility with Cluster Shapes: Hierarchical clustering can capture clusters of different shapes and sizes better than KMeans, which assumes spherical clusters.

-Deterministic Output: The output of hierarchical clustering is deterministic, whereas KMeans can produce different clusters on different runs of the algorithm due to random initialization of centroids.


2.Advantages of Density-Based Clustering (like DBSCAN) over KMeans:

-Ability to Handle Noise: Density-based clustering methods can identify and handle noise points (outliers), which KMeans cannot do effectively.

-No Assumption of Spherical Clusters: Unlike KMeans, density-based clustering does not assume that clusters are spherical.

-It can find arbitrarily shaped clusters.

-Does Not Require Specifying Number of Clusters: It does not require the number of clusters to be specified in advance.

-Robust to Initialization: It doesnt rely on the initial placement of centroids or seeds, thus overcoming a significant limitation of KMeans related to initialization sensitivity.

-Capability to Discover Clusters of Varying Densities: Density-based methods can find clusters of varying densities, which KMeans, with its equal variance assumption, might miss.

The dendrogram produced can be used to decide the number of clusters by cutting the dendrogram at a suitable level. K-means requires the number of clusters to be specified beforehand, which can be challenging to estimate correctly without prior knowledge.

Flexibility in Cluster Shapes: Hierarchical clustering can accommodate clusters of different shapes and sizes much better than K-means, which assumes clusters to be spherical and tends to work well when the clusters are of similar size.

Understanding Data Structure: The dendrogram from hierarchical clustering provides a deep insight into the datas structure, revealing at a glance the way individual elements are grouped into clusters. This can be particularly useful for exploratory data analysis.

Quality of Clustering: Hierarchical clustering can sometimes produce more accurate clusters than K-means. Especially in cases where the data doesnt conform to the assumptions made by K-means (such as clusters of equal variance and spherical shape), hierarchical clustering can be more reliable.

Robustness to the Initial Conditions: K-means clustering is sensitive to the initial placement of centroids and can converge to local optima. This means it can produce different results on different runs if the initial centroids change. Hierarchical clustering doesnt have this issue as it doesnt rely on the initialization of centroids.

Versatility in Distance Measures: Hierarchical clustering can work with any distance metric (like Euclidean, Manhattan, cosine similarity, etc.), whereas K-means typically uses Euclidean distance. This makes hierarchical clustering more versatile for different types of data and use cases.

However, it's important to note that hierarchical clustering is computationally more expensive than K-means, especially for large datasets, as it typically requires computing and storing distances between all pairs of points. Therefore, the choice between hierarchical clustering and K-means often depends on the specific requirements of the task at hand, including the nature of the data, the computational resources available, and the specific objectives of the clustering exercise.

2.Density-Based Clustering, with DBSCAN (Density-Based Spatial Clustering of Applications with Noise) being one of the most popular algorithms in this category, offers several advantages over K-means, especially in certain contexts and for specific types of data. Here are some of the key advantages:

Handling Noise: One of the main advantages of DBSCAN and similar density-based methods is their ability to identify and handle noise or outliers effectively. Points in low-density regions are treated as noise and excluded from clusters, which can lead to more meaningful clustering results. K-means, on the other hand, assigns every point to a cluster, even if it is an outlier.

Cluster Shape Flexibility: Density-based clustering does not assume any particular shape for the clusters, unlike K-means, which implicitly assumes that clusters are spherical. Therefore, DBSCAN can find arbitrarily shaped clusters, which is useful in many real-world scenarios where data may not be spherical or well-separated.

Number of Clusters: Unlike K-means, DBSCAN does not require specifying the number of clusters in advance. The algorithm determines the number of clusters based on the input data and the parameters for density ( - epsilon and minPts). This feature makes DBSCAN more useful in situations where the number of clusters is not known beforehand.

Handling Clusters of Different Densities: DBSCAN can find clusters of varying densities, which is a significant advantage over K-means. K-means can struggle with clusters of varying densities and sizes since it tends to partition the data space into roughly equal parts.

Robustness to Initialization: K-means is sensitive to the initial placement of centroids and can converge to different local optima depending on the initial state. This isn't an issue with DBSCAN, as it does not rely on such initialization.

No Assumption of Homogeneity in Cluster Sizes or Densities: K-means works best when all clusters are of approximately the same size and density. DBSCAN, however, performs well even when this is not the case.

Scalability with Large Datasets: Recent variations of DBSCAN (like HDBSCAN - Hierarchical DBSCAN) are designed to be more scalable with large datasets, making them suitable for big data applications.

Despite these advantages, DBSCAN and other density-based methods also have limitations. Choosing appropriate values for  and minPts can be challenging and can significantly affect the outcome. Also, DBSCAN's performance can degrade with high-dimensional data due to the curse of dimensionality. Therefore, the choice between DBSCAN and K-means should be based on the specific requirements and characteristics of the data at hand.
